```python
def definir_plataforma(filme):
  netflix = filme['Netflix']
  hulu = filme["Hulu"]
  prime = filme["Prime Video"]
  disney = filme["Disney+"]

  if netflix == 1:
    return "Netflix"
  elif hulu == 1:
    return "Hulu"
  elif prime == 1:
    return "Prime Video"
  elif disney == 1:
    return "Disney+"
  elif netflix !=1 & hulu != 1 & prime != 1 & disney != 1:
    return "NoPlataform"
```

```python
#Testee
teste = df.iloc[0]
teste
teste['Type'] = definir_plataforma(teste)
teste
```

```python
df['Type'] = df.apply(definir_plataforma, axis=1)
```

###Melhores plataformas por Decada

```python
def hist_plataforma(df):
  # Configurando tema e estilo
  sns.set_theme(style="whitegrid")

  # Ajustando a paleta de cores e o gráfico
  plt.figure(figsize=(10, 6))
  sns.histplot(data=df, x="Type", multiple="stack", palette="Set2", shrink=0.8)

  # Adicionando rótulos e título
  plt.xlabel('Tipo de Plataforma', fontsize=12)
  plt.ylabel('Contagem', fontsize=12)
  plt.title(f"Distribuição de Tipos de Plataformas no ano de {df['Decade']}", fontsize=14)

  # Rotacionando os rótulos do eixo X
  plt.xticks(rotation=45, ha='right')

  # Exibindo o gráfico
  plt.tight_layout()
  plt.show()
```

```python
y_min = df.groupby("Decade")['Type'].count().min()
y_max = df.groupby("Decade")['Type'].count().max()
y_limits = (y_min, y_max)
```

```python
def hist_decada(df_plataforma):
  # Configurando tema e estilo
  sns.set_theme(style="whitegrid")
  plataforma = df_plataforma.iloc[0]['Type']
  bins = list(df_plataforma['Decade'].unique())
  bins.sort()
  # Ajustando a paleta de cores e o gráfico
  plt.figure(figsize=(10, 8))
  if plataforma == 'Netflix':
    color="#E50914"
    label='Netflix'
  elif plataforma == 'Hulu':
    color="#8BC34A"
    label='Hulu'
  elif plataforma == 'Prime Video':
    color="#29B6F6"
    label='Prime Video'
  elif plataforma == 'Disney+':
    color="#0D61A9"
    label='Disney+'

  sns.histplot(data=df_plataforma, x="Decade",  color=color, shrink=1,bins=bins,label=label)


  # Adicionando rótulos e título
  plt.xlabel('Tipo de Plataforma', fontsize=12)
  plt.ylabel('Contagem', fontsize=12)
  plt.title(f"Distribuição da Plataforma {plataforma} por decada", fontsize=14)
  y_min = 0
  y_max = 4000
  y_limits = (y_min, y_max)
  plt.ylim(y_limits)

  # Rotacionando os rótulos do eixo X
  plt.xticks(rotation=45, ha='right')

  plt.tight_layout()
  plt.show()
```

```python
for plataform in df['Type'].unique():
  df_plataforma = df[df['Type'] == plataform]
  hist_decada(df_plataforma)
```
